```python
from imageai.Detection import ObjectDetection
```

```python
detector = ObjectDetection()
```

```python
import urllib.request

urllib.request.urlretrieve(
    'https://gitlab.com/andreass.bayu/imageai-repo/-/raw/master/resnet50_coco_best_v2.1.0.h5', 'model h5')
```

```python
import urllib.request

urllib.request.urlretrieve(
    'https://gitlab.com/andreass.bayu/imageai-repo/-/raw/master/object-detection-with-imageai-python-1.jpg', 'image-2.jpg')
```

```python
model_path = "model h5"
input_path = "image-2.jpg"
output_path = "newimage.jpg"
```

```python
detector.setModelTypeAsRetinaNet()
```

```python
detector.setModelPath(model_path)
```

```python
detector.loadModel()
```

```python
detection = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)
```

```python
for eachItem in detection:
  print(eachItem["name"], ":", eachItem["percentage_probability"]) 
```
