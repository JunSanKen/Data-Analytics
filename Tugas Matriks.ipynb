```python
#Penambahan Dua Matriks

mata =[
    [3, 4],
    [2, 1]
]
matb =[
    [1, 5],
    [3, 7]
]
result =  [[0,0],
          [0, 0]]
for x in range(len(mata)):

    for y in range(0, len(mata[0])):
      result[x][y]+= mata [x][y]+matb [x][y]
        
for r in result:
  print(r)
```

```python
#Perkalian Dua Matriks

mata =[
    [3, 4],
    [2, 1]
]
matb =[
    [1, 5],
    [3, 7]
]

matc = [
]

result =  [[0,0],
          [0, 0]]
for x in range(0, len(mata)):
    row = []
    for y in range(0, len(mata[0])):
        total = 0
        for z in range(0, len(mata)):
            total = total + (mata[x][z] * matb[z][y])
        row.append(total)
    matc.append(row)

for x in range(0, len(matc)):
    for y in range(0, len(matc[0])):
      print ([matc[x][y]], end =' ')
    print ()
        
```

```python
#Transpos Matriks

mata = [[3, 4],
        [2,1]]

result = [[0, 0],
         [0, 0]]

for x in range(len(mata)):
    for y in range (len(mata[0])):
      result[x][y] = mata[x][y]

for r in result:
  print (r)
```
