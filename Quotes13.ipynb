```python
# Library yang perlu diimport
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "https://quotes.toscrape.com/"
response = requests.get(url)
response = response.content
soup = BeautifulSoup(response, "html.parser")
articles = soup.find_all('div', class_='quote')
print(articles)
```

```python
kutipan = []
for page in range(1, 15):
  url = "https://quotes.toscrape.com/page/" + str(page)
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, "html.parser")
  quotes = soup.find_all('div', class_='quote')
```

```python
for quote in quotes:
    kutip = quote.find('div', attrs={'class': 'text'}).text
    author = quote.find('div', attrs={'class': 'author'}).text
    tags = quote.find('div', attrs={'class': 'tag'}).text
```
