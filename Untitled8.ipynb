```python
## PROGRAM MP3 PLAYER ##
## TPT - KELOMPOK 5 - 2KA01 ##

# Import library yang akan digunakan
import sys
import os
from tkinter.font import Font
import pywhatkit
import easygui
import qdarkgraystyle

# Import widget-widget yang digunakan dari library
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from PyQt5.QtMultimedia import QMediaContent, QMediaPlayer

class MP3Player(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Music Player")
        self.setWindowIcon(QIcon("D:\TPT_Python\Music Player\music.png"))
        # Mengecek kondisi lagu dimainkan
        self.state = "Play"
        self.playlist = []
        self.position = 0
        self.index = ""
        self.init_ui()
        
# LAYOUT WINDOW MUSIC PLAYER 
    def init_ui(self):
        vb = QVBoxLayout()
        self.setLayout(vb)
        vb.setAlignment(Qt.AlignCenter)
        
        self.label = QLabel("MP3 Player")
        self.label.setFont(QFont("OCR A STD", 20))
        self.label.setAlignment(Qt.AlignCenter)
        vb.addWidget(self.label)
        
        hb = QHBoxLayout()
        vb.addLayout(hb)
        
        #Tombol-tombol
        font = QFont("OCR A STD", 10)
        
        # Tombol Memindahkan Lagu ke Sebelumnya
        self.skipbackbtn = QPushButton()
        self.skipbackbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaSkipBackward))
        hb.addWidget(self.skipbackbtn)
        
        # Tombol Memundurkan Lagu Beberapa Detik
        self.backwardbtn = QPushButton()
        self.backwardbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaSeekBackward))
        self.backwardbtn.setFont(font)
        hb.addWidget(self.backwardbtn)
        
        # Tombol Play Lagu
        self.playbtn = QPushButton()
        self.playbtn.setEnabled(False)
        self.playbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaPlay))
        self.playbtn.setFont(font)
        hb.addWidget(self.playbtn)
        
        # Tombol Memajukan lagu beberapa detik
        self.forwardbtn = QPushButton()
        self.forwardbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaSeekForward))
        self.forwardbtn.setFont(font)
        hb.addWidget(self.forwardbtn)
        
        # Tombol Next (pindah lagu ke lagu berikutnya)
        self.skipnextbtn = QPushButton()
        self.skipnextbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaSkipForward))
        hb.addWidget(self.skipnextbtn)
        
    # Layout baris berikutnya
        hb2 = QHBoxLayout()
        vb.addLayout(hb2)
        
        # Tombol untuk buka folder
        self.openfilebtn = QPushButton()
        self.openfilebtn.setIcon(self.style().standardIcon(QStyle.SP_DirOpenIcon))
        self.openfilebtn.setFont(font)
        hb2.addWidget(self.openfilebtn)
        
        # Membuat tampilan slider lagu
        self.slider = QSlider(Qt.Horizontal)
        self.slider.setRange(0,0)
        self.slider.sliderMoved.connect(self.set_position)
        hb2.addWidget(self.slider)
        
    # Playlist
        self.songlist = QListWidget()
        self.songlist.setFont(font)
        vb.addWidget(self.songlist)
        
        
    # Toolbar
        self.toolbar = QToolBar()
        vb.addWidget(self.toolbar)
        
        # Tombol Search Lagu dari YouTube
        self.searchbtn = QPushButton()
        self.searchbtn.setText('YT Search')
        self.searchbtn.setIcon(self.style().standardIcon(QStyle.SP_ComputerIcon))
        self.searchbtn.setFont(font)
        self.toolbar.addWidget(self.searchbtn)
        self.toolbar.addSeparator()
        
        # Untuk memutar lagu
        self.player = QMediaPlayer()
        
        self.search_lyrics_btn = QPushButton()
        self.search_lyrics_btn.setFont(font)
        self.search_lyrics_btn.setText('Lyrics')
        self.toolbar.addWidget(self.search_lyrics_btn)
        self.toolbar.addSeparator()
        
        self.space = QWidget()
        self.space.setSizePolicy(QSizePolicy.Expanding, QSizePolicy.Preferred)
        self.toolbar.addWidget(self.space)
        
        
        # Speaker dan Volume
        self.speaker = QAction() 
        self.speaker.setIcon(self.style().standardIcon(QStyle.SP_MediaVolume))
        self.toolbar.addAction(self.speaker)
        self.toolbar.addAction(self.speaker)
        self.slider_v1 = QSlider(Qt.Horizontal)
        self.toolbar.addWidget(self.slider_v1)
        self.label_v1 = QLabel("50%")
        self.label_v1.setFont(QFont("OCR A STD", 10))
        self.label_v1.setMinimumWidth(20)
        self.toolbar.addWidget(self.label_v1)
        
        self.slider.setRange(0,100)
        volume = self.player.volume()
        self.slider_v1.setValue(int(volume/2))
        
    # Koneksi tombol dengan fungsi tombol
        # Memilih dari folder, memilih dari playlist dan memutarnya
        self.openfilebtn.clicked.connect(self.open_mp3_file)
        self.playbtn.clicked.connect(self.play_mp3)
        
        # Memanggil fungsi untuk slider agar bergerak sesuai durasi lagu
        self.player.positionChanged.connect(self.position_changed)
        self.player.durationChanged.connect(self.duration_changed)

        # memanggil fungsi pengubah icon play/pause
        self.player.stateChanged.connect(self.state_changed)
        
        # memanggil fungsi tombol dan lagu pada playlist
        self.forwardbtn.clicked.connect(self.move_forward)
        self.backwardbtn.clicked.connect(self.move_backward)
        self.songlist.clicked.connect(self.set_state) #Agar lagu yang diklik dari playlist dapat diputar
        self.songlist.doubleClicked.connect(self.play_mp3) #Memutar lagu pada playlist yang di klik 2x
        self.skipbackbtn.clicked.connect(self.skip_backward)
        self.skipnextbtn.clicked.connect(self.skip_next)
        self.searchbtn.clicked.connect(self.search_from_yt)
        self.search_lyrics_btn.clicked.connect(self.search_lyrics)
        self.slider_v1.valueChanged.connect(self.set_volume)
    # Fungsi untuk membuka file mp3
    def open_mp3_file(self):
        file_name = QFileDialog()
        file_name.setFileMode(QFileDialog.ExistingFiles)
        names = file_name.getOpenFileNames(self, "Open Files", os.getenv("HOME"))
        self.song = names[0]
        self.songlist.addItems(self.song)
        
    # Fungsi untuk mengetahui kondisi lagu    
    def set_state(self):
        self.playbtn.setEnabled(True)
        self.state = "Play"
        self.playbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaPlay))
            
    # Fungsi untuk memutar lagu
    def play_mp3(self):
        #kondisi jika lagu keadaan 'Play'
        if self.state == "Play":
            self.playbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaPlay))
            self.state = "Pause"
            
            path = self.songlist.currentItem().text()
            url = QUrl.fromLocalFile(path)
            content = QMediaContent(url)
            self.player.setMedia(content)
            self.index = self.songlist.currentRow().__index__()
            
            self.player.setPosition(self.position)
            self.playlist.append(path)
        
        #Fungsi pada tombol play/pause agar lagu tidak dimulai dari awal setalah di pause    
            if len(self.playlist) > 2:
                self.playlist.pop(0)
            if self.songlist.currentItem().text() != self.playlist[0]:
                self.position = 0
                self.player.setPosition(self.position)
                
            self.player.play()
        # kondisi jika lagu sudah dalam keadaaan 'Pause'
        else:
            self.state = "Play"
            self.player.pause()
            paused = self.player.position()
            self.position = paused
    

    # Fungsi untuk memindahkan lagu ke lagu sebelumnya
    def skip_backward(self):
        self.state = "Play"
        try:
            self.songlist.setCurrentRow(self.index - 1)
            self.play_mp3()
        except:
            pass
    
    # Fungsi untuk memindahkan lagu ke lagu berikutnya    
    def skip_next(self):
        self.state = "Play"
        try: 
            self.songlist.setCurrentRow(self.index + 1)
            self.play_mp3()       
        except:
            pass
    
    
    def set_position(self, position):
        self.player.setPosition(position)
    
    def position_changed(self, position):
        self.slider.setValue(position)
        duration = self.player.duration()
        value = self.slider.value()
        if value == duration:
            self.state = "Play"
            self.play_mp3()
    
    # Membuat slider bergerak sesuai durasi lagu        
    def duration_changed(self, duration):
        self.slider.setRange(0, duration) 
        
    # Mengubah icon play menjadi pause jika lagu di pause
    # dan icon pause menjadi play saat lagu diputar kembali setelah di pause
    def state_changed(self, state):
        if self.player.state() == QMediaPlayer.PlayingState:
            self.playbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaPause))
        else:
            self.playbtn.setIcon(self.style().standardIcon(QStyle.SP_MediaPlay))
    
    # Fungsi untuk maju beberapa detik pada lagu yang sedang diputar
    def move_forward(self):
        self.player.setPosition(int(self.player.position()) + 2000)
    
    # Fungsi untuk mundur pada lagu yang diputar beberapa detik
    def move_backward(self):
        self.player.setPosition(int(self.player.position()) - 2000)
        
    
    # Fungsi untuk mencari lagu dari YouTube
    def search_from_yt(self):
        e = easygui.enterbox("Enter the song title", "Youtube Search")
        if e :
            pywhatkit.playonyt(e)
    
    def search_lyrics(self):
        l = easygui.enterbox("Enter the song title", "Find Song Lyrics")
        if l:
            pywhatkit.search(l + ' lyrics')
                    
    # Fungsi untuk mengatur volume
    def set_volume(self):
        volume = self.slider_v1.value()
        self.player.setVolume(volume)
        self.label_v1.setText(str(volume)+"%")
    
               
# Tampilan Window        
def main():
    app = QApplication(sys.argv)
    app.setStyleSheet(qdarkgraystyle.load_stylesheet())
    gui = MP3Player()
    gui.setGeometry(500, 500, 500, 500)
    gui.show()
    sys.exit(app.exec_())
    
if __name__ == '__main__':
    main()
    
    
    
```
