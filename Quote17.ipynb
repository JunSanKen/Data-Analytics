```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "http://quotes.toscrape.com/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
else:
    print("Error fetching the website")
```

```python
quotes = []
for quote_element in soup.find_all(class_="quote"):
    quote = quote_element.find("span").get_text()
    author = quote_element.find("small").get_text()
    quotes.append({"quote": quote, "author": author})
```

```python
with open("quotes.csv", "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["quote", "author"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(quotes)
```
