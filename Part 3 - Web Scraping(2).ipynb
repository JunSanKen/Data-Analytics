```python
# Library yang perlu diimport
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
root = "https://www.google.com"
link = "https://www.google.com/search?q=pemilihan+presiden+2024&sca_esv=593582740&rlz=1C5CHFA_enID988ID988&biw=1440&bih=677&tbm=nws&sxsrf=AM9HkKnLd1todjWqOHlMlB654I_ufaW8Sg%3A1703509245101&ei=_XyJZY_mBY-gseMPhMuHiAY&oq=Pemilihan+Pres&gs_lp=Egxnd3Mtd2l6LW5ld3MiDlBlbWlsaWhhbiBQcmVzKgIIATIOEAAYgAQYgAQYsQMYgwEyCxAAGIAEGIAEGLEDMgsQABiABBiABBixAzIIEAAYgAQYgAQyDhAAGIAEGIAEGLEDGIMBMggQABiABBiABDIIEAAYgAQYgAQyCBAAGIAEGIAEMg4QABiABBiABBixAxiDATIIEAAYgAQYgARI9iVQAFjQFXAAeACQAQGYAe8BoAGxC6oBBjExLjIuMbgBA8gBAPgBAcICCBAAGIAEGLEDwgILEAAYgAQYsQMYgwHCAg4QABiABBiKBRixAxiDAcICBRAAGIAEwgINEAAYgAQYgAQYigUYQ8ICERAAGIAEGIAEGIoFGLEDGIMBiAYB&sclient=gws-wiz-news"
def news(link):
    global root

    session = requests.Session()
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = session.get(link, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    print(soup)
```

```python
news_results = soup.find_all('div', attrs={'class': 'Gx5Zad fP1Qef xpd EtOod pkphOe'})
#     print(news_results)

    data = []

    for item in news_results:
        judul = item.find('div', attrs={'class': 'BNeawe vvjwJb AP7Wnd'}).get_text()
        penerbit = item.find('div', attrs={'class': 'BNeawe UPmit AP7Wnd lRVwie'}).get_text()
        deskripsi = item.find('div', attrs={'class': 'BNeawe s3v9rd AP7Wnd'}).get_text()
        waktu = deskripsi.split(".")[-1]
        raw_link = (item.find('a', href=True)['href'])
        link = (raw_link.split("/url?q=")[1]).split('&sa=U&')[0]
#         print(judul)


        data.append({'Judul': judul, 'Penerbit': penerbit, 'Waktu': waktu, 'Link' : link})

    next_page = soup.find('a', attrs={'aria-label': 'Halaman berikutnya'})

    if next_page:
        next_page_url = root + next_page['href']
        data.extend(news(next_page_url))

    return data

data = news(link)

# Save the data to a CSV file
csv_file = 'berita_ds.csv'
csv_columns = ['Judul', 'Penerbit', 'Waktu', 'Link']
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=csv_columns)
    writer.writeheader()
    writer.writerows(data)

print("Data saved to", csv_file)
```
