## Data Preparation

### Data Cleansing

```python
import numpy as np
```

```python
## Handling Outliers (extreme values)
q1 = np.quantile(df['LotArea'].dropna(), 0.25)
q3 = np.nanquantile(df['LotArea'], 0.75)
iqr = q3 - q1

upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
print(q1, q3, iqr, lower_bound, upper_bound)
```

```python
df['LotArea'].describe()
```

```python
# drop columns with missing values
df_no_nan = df.dropna()
# df['FoodCourt'] = df['FoodCourt'].dropna()
print(df.shape, df_no_nan.shape)
```

```python
## rule of thumbs, if missing values is not more than 40% than fill it

# fill using some defined values
df['Alley'] = df['Alley'].fillna("No Info")
df['PoolQC'] = df['PoolQC'].fillna(0.)
```
